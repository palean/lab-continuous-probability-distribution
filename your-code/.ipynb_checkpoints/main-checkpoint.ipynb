## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
from scipy.stats import uniform
x = uniform.rvs(size=10)
a = 2
b = 3
randoms  = a + (b-a)*x
print(randoms)
```

```python
import numpy as np

# Numpy way
randoms_np = np.random.uniform(2,3,(1,10)) # (low, high, size)
                                           # lower boundary, upper boundary, output shape
print(randoms_np.tolist()[0])
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
# your code here
import numpy as np

uniforme = lambda bottom, ceiling, count: np.random.uniform(bottom, ceiling, count)
uniforme(10,15,100)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
%matplotlib inline

plt.rcParams['figure.figsize'] = (22,12)
fig, axes=plt.subplots(1,2,sharey=True)
plt.suptitle('Uniform Distribution', size=25)

axes[0].hist(uniforme(10,15,100));
axes[1].hist(uniforme(10,60,1000));

# For y-x-axis sizes
axes[0].tick_params(labelsize=18)
axes[1].tick_params(labelsize=18)

#Shared axis
fig.text(0.5,0.05, "x coordenate", ha="center", va="center", fontsize=20)
fig.text(0.05,0.5, "Uniform Distribution", ha="center", va="center", fontsize=20, rotation=90)

# To answer: "How are the two distributions different?"

# build a rectangle in axes coords
left, width = .15, .75
bottom, height = .25, .75
right = left + width
top = bottom + height

axes[0].text(right, top, 'This graphic has a range between 10 to 15, with a sample size of 100 \n',
        horizontalalignment='right',
        verticalalignment='bottom',
        transform=axes[0].transAxes, size=18)
axes[1].text(right, top, 'Whereas the second one has a widely range from 10 to 60, with a numerous sample of 1000.',
        horizontalalignment='right',
        verticalalignment='bottom',
        transform=axes[1].transAxes, size=18);
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
# your code here
fig, axes=plt.subplots(1,2,sharey=True)
plt.suptitle('Normal Distribution', size=25)

axes[0].hist(np.random.normal(10,1,1000), bins=100, color='r');
axes[1].hist(np.random.normal(10,50,1000), bins=100, color='r');

axes[0].tick_params(labelsize=18)
axes[1].tick_params(labelsize=18)
```

How are the two distributions different?

```python
# your answer below
# They are different in its standard desviation around its average. 
#The standard deviation of the second graphic is 50 and it is more widly distributed 
#from the center than the sample with a standard deviation of 1 from the first graphic.

# So, for the first graphic, the 95% of the data is located between 9 and 11,
#whereas on the second figure, the 95% of the data is located in a bigger range 
# from 40 to 60.
```

## Normal Distribution of Real Data

In this challenge we are going to take a look the real data. We will use vehicles.csv file for this exercise

First import vehicles.csv.
Then plot the histograms for the following variables:

1. Fuel Barrels/Year

```python
# your code here
import pandas as pd
import scipy.stats as stats

data = pd.read_csv('vehicles.csv')
data.head(3)

fig, axes=plt.subplots(1,3,sharey=True)

data['Fuel Barrels/Year'].hist(bins=50,ax=axes[0]);
data['CO2 Emission Grams/Mile'].hist(bins=50, ax=axes[1]);
data['Combined MPG'].hist(bins=50, ax=axes[2]);
```

1. Fuel Barrels per Year

```python
#Compaired to a normal distribution
import seaborn as sns

#Data
sns.distplot(data['Fuel Barrels/Year'], label='Real data', color='r')
a = data['Fuel Barrels/Year'].mean()
b = data['Fuel Barrels/Year'].std()
sns.distplot(np.random.normal(a,b,10000),label='Predicted data')

#General format
plt.title('Distribution of fuel barrels per year')
sns.set_style('darkgrid')
plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.labelsize'] = 20
plt.rcParams.update({'font.size':20})

plt.legend()
plt.show()
```

2. CO2 Emission Grams/Mile 

```python
#Data
sns.distplot(data['CO2 Emission Grams/Mile'], label='Real data', color='r')
a = data['CO2 Emission Grams/Mile'].mean()
b = data['CO2 Emission Grams/Mile'].std()
sns.distplot(np.random.normal(a,b,10000),label='Predicted data')

#General format
plt.title('Distribution of CO2 Emission Grams per Mile')

plt.legend()
plt.show()
```

3. Combined MPG

```python
#Data
sns.distplot(data['Combined MPG'], label='Real Data', color='r')
a= data['Combined MPG'].mean()
b= data['Combined MPG'].std()
sns.distplot(np.random.normal(a,b,10000), label='Predicted data')

#Format
plt.title('Distribution of Combined MPG')

plt.legend()
plt.show()
```

Which one(s) of the variables are nearly normally distributed? How do you know?

```python
# Applying normal test that return True or False to know if they are nromal distributed data

print(stats.normaltest(data['Fuel Barrels/Year'])[1]<1e3)
print(stats.normaltest(data['CO2 Emission Grams/Mile'])[1]<1e3)
print(stats.normaltest(data['Combined MPG'])[1]<1e3)

#All of them can be considered as normal distributions
```

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e.  = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
plt.hist(np.random.exponential(10,1000), bins=100, color='g')
plt.show()
```

```python
# your answer here
# Hint: This is same as saying P(x<15)
distr = np.random.exponential(10,1000)
area = len([i for i in distr if i<=15])/10
area
```

What is the probability that the customer will spend more than 15 minutes

```python
print('The probability that the customer will spend more than 15 mins is',sum(distr>=15)/10,'%')
```
